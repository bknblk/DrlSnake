```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('training_log.csv')
df.head()
```

```python
plt.figure()
plt.plot(df["episode"], df["reward"])
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Reward per Episode")
plt.grid(True)
plt.savefig("Reward per Episode.png")
plt.show()
```

```python
plt.figure()
plt.plot(df["episode"], df["steps"])
plt.xlabel("Episode")
plt.ylabel("Steps")
plt.title("Steps Survived per Episode")
plt.grid(True)
plt.savefig("Steps Survived per Episode.png")
plt.show()
```

```python
plt.figure()
plt.plot(df["episode"], df["loss"])
plt.xlabel("Episode")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(df["episode"], df["reward"], label="Reward")
plt.plot(df["episode"], df["steps"], label="Steps")
plt.legend()
plt.xlabel("Episode")
plt.title("Reward and Steps Over Training")
plt.grid(True)
plt.show()
```

```python
window = 5
smooth_reward = df["reward"].rolling(window).mean()

plt.figure()
plt.plot(df["episode"], df["reward"], alpha=0.3, label="Raw")
plt.plot(df["episode"], smooth_reward, label=f"{window}-episode MA")
plt.legend()
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Smoothed Reward Curve")
plt.grid(True)
plt.savefig("Smoothed Reward Curve.png")
plt.show()
```
